# WEB SCARPING AND DATA CLEANING

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
link='https://www.worldometers.info/coronavirus/'
r=requests.get(link)
soup=BeautifulSoup(r.text,'html.parser')
```

```python
x=soup.find('tbody')
```

```python
country=[]
total_cases=[]
new_cases=[]
total_deaths=[]
new_deaths=[]
total_recovered=[]
new_recovered=[]
active_cases=[]
serious=[]
TopCases_1MPop=[]
Death_1MPop=[]
total_tests=[]
Tests_1MPop=[]
population=[]
for row in x.find_all('tr'):
    col=row.find_all('td')
    country.append(col[1].text)
    total_cases.append(col[2].text)
    new_cases.append(col[3].text)
    total_deaths.append(col[4].text)
    new_deaths.append(col[5].text)
    total_recovered.append(col[6].text)
    new_recovered.append(col[7].text)
    active_cases.append(col[8].text)
    serious.append(col[9].text)
    TopCases_1MPop.append(col[10].text)
    Death_1MPop.append(col[11].text)
    total_tests.append(col[12].text)
    Tests_1MPop.append(col[13].text)
    population.append(col[14].text)
```

```python
data=pd.DataFrame({'Country':country,'Total Cases':total_cases,'New Cases':new_cases,'Total Deaths':total_deaths,'New Deaths':new_deaths,'Total Recovered':total_recovered,'New Recovered':new_recovered,'Active Cases':active_cases,'Serious':serious,'Top Cases/1MPop':TopCases_1MPop,'Deaths/1MPop':Death_1MPop,'Total Tests':total_tests,'Tests/1MPop':Tests_1MPop,'Population':population})
data
```

```python
data['Country'] = data['Country'].str.replace('\n', '')
data['Country']=data['Country'].replace('',np.nan)
data=data.dropna(subset=['Country'])
for col in data.columns[1:]:
    data[col] = data[col].astype(str)
    data[col] = data[col].str.strip()
    data[col] = data[col].str.replace(',', '')
    data[col] = data[col].str.replace('+', '')
    data[col] = pd.to_numeric(data[col], errors='coerce')
data.head(10)
```

```python
data.isna().sum()
```

```python
data['New Cases'].fillna(0, inplace=True)
data['Total Deaths'].fillna(data['Total Deaths'].median(), inplace=True)
data['New Deaths'].fillna(0, inplace=True)
data['Total Recovered'].fillna(data['Total Recovered'].mean(), inplace=True)
data['New Recovered'].fillna(0, inplace=True)
data['Active Cases'].fillna(data['Active Cases'].mean(), inplace=True)
data['Serious'].fillna(0, inplace=True)
data['Top Cases/1MPop'].fillna(data['Top Cases/1MPop'].mean(), inplace=True)
data['Deaths/1MPop'].fillna(data['Deaths/1MPop'].mean(), inplace=True)
data['Total Tests'].fillna(data['Total Tests'].mean(), inplace=True)
data['Tests/1MPop'].fillna(data['Tests/1MPop'].mean(), inplace=True)
data['Population'].fillna(data['Population'].mean(), inplace=True)
```
